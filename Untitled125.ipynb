# ***Handling null values***
a. For categorical variables with null values, update those rows as
unknown_column_name.

```python
df.isnull().sum()
```

# Drop the null values which are less than 5% from the entire data

```python
df.dropna(subset = 'date_added',inplace = True)
df.dropna(subset = 'rating',inplace = True)
df.dropna(subset = 'duration',inplace = True)
```

```python
df.isnull().sum()
```

```python
df['director'].replace(np.nan,'No director',inplace = True)
df['cast'].replace(np.nan,'No cast',inplace = True)
df['country'].replace(np.nan,'No country',inplace = True)
```

```python
df.isnull().sum()
```

```python
df.head(5)
```

**seperating the time duration in the duration columns **

```python
df['duration']
```

```python
df['movie_duration'] = df['duration'].where(df['duration'].str.contains('min'),'')
df['season_duration'] = df['duration'].where(df['duration'].str.contains('Seasons','Season'),'')
df.drop('duration',axis = 1 ,inplace = True)
```

```python
df.head()
```

```python
df['date_added'].dt.month
```

```python
df['date_added'] = pd.to_datetime(df['date_added'],errors = 'coerce')
```

```python
df['month'] = df['date_added'].dt.month
df['year'] = df['date_added'].dt.year
```

```python
df.head()
```

```python
df['cast']
```

# ***Un-nesting the columns***
a. Un-nest the columns those have cells with multiple comma separated values by
creating multiple rows

```python
df['cast'] = df['cast'].str.split(',')
df = df.explode('cast')
df.reset_index(drop = True)
```

```python
df.index = np.arange(1,len(df)+1)
```

```python
df.head()
```

```python
df['listed_in'] = df['listed_in'].str.split(',')
df = df.explode('listed_in')
df.reset_index(drop = True)
```

```python
df.index = np.arange(1,len(df)+1)
```

```python
df.head()
```

# ***Data Visualization***

Comparison of tv shows vs. movies.

```python
plt.figure(figsize = (6,5))
sns.countplot(x= 'type' ,data = df ,width = 0.4,color = 'red')
plt.xlabel('TYPE',fontsize = 12 )
plt.ylabel('COUNT',fontsize = 10)
plt.title('MOVIE VS TV_SHOW')
plt.show()
```


# **Top 10 Directors by the Number of Films and TV Shows They Directed**

```python
director_count = df.groupby('director')['type'].count().sort_values(ascending = False)[1:11]
```

```python
plt.figure(figsize = (12,8))
sns.countplot(x = 'director' , data = df ,order = director_count.index)
plt.title('TOP 10 DIRECTOR',fontsize = 15)
plt.xlabel('director',fontsize = 12 )
plt.ylabel('count',fontsize = 12 )
plt.xticks(rotation = 90 ,size = 12)
plt.show()
```

 ***Find the number of movies produced in each country and pick the top 10 countries***


```python
top_10_country = df.groupby('country')['director'].count().sort_values(ascending = False)[1:11]
top_10_country
```

```python
plt.figure(figsize = (12,8))
sns.countplot(x = 'country' , data = df ,order = top_10_country.index,color = 'green')
plt.title('Top 10 Country',fontsize = 15)
plt.xlabel('Country',fontsize = 12 )
plt.ylabel('count',fontsize = 12 )
plt.xticks(rotation = 90 ,size = 12)
plt.show()
```

# ***What is the best time to launch a TV show?***
a. Find which is the best week to release the Tv-show or the movie. Do the analysis
separately for Tv-shows and Movies

```python
df['week_number'] = df['date_added'].dt.isocalendar().week
```

```python
weekno = df.groupby('week_number')['type'].count().sort_values(ascending = False)[:5]
```

```python
sns.countplot(x = 'week_number',data = df , order = weekno.index)
```

```python
week_number = 26
year = 2008
first_day_of_year = pd.to_datetime(str(year), format='%Y')
first_monday = first_day_of_year + pd.DateOffset(weekday=0)
date_from_week = first_monday + pd.DateOffset(weeks=week_number - 1)
print("Date from week number {}: {}".format(week_number, date_from_week))
```
